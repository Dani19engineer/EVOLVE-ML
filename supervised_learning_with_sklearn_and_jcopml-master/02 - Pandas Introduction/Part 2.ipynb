# Map, Apply & Transform

```python
df["shelf_str"]=df.shelf.map({1: "bawah", 2:"tengah", 3: "atas"})
df.head()
```

```python
def konversi_ke_mg(x):
    return x*1000

def konversi_jadi_huruf_kecil(x):
    return x.lower()
```

```python
df['fat_mg'] = df.fat.apply(konversi_ke_mg)
df.head()
```

```python
df['mfr'] = df.mfr.apply(konversi_jadi_huruf_kecil)
df.head()
```

```python
df['protein_mg'] = df.protein.apply(lambda x:x*1000)
df.head()
```

```python
df['type_lower'] = df.type.apply(lambda x: x.lower())
df.head()
```

```python
df['rating'] = df.rating.transform(lambda x: x-x.mean())
df.head()
```

apply itu melakukan fungsi ke saty per satu (tidak bisa ambil informasi keseluruhan seperti mean), tapi transform bisa.
