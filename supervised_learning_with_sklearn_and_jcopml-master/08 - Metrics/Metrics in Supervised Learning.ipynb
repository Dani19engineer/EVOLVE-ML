# Regression Metrics

### R-squared

Seberapa bagus model kita dibandingkan rata-rata semua. Berkisar dari -inf sampai 1.<br>
Negatif artinya lebih buruk dari menggunakan rata-rata sederhana.<br>
0 artinya sama baiknya dengan rata-rata.
Positif artinya lebih baik dari rata-rata, dan 1 artinya error = 0

### Mean Absolute Error (MAE)

Rata-rata dari error yang dimutlakkan. Semakin kecil semakin baik.

### Mean Squared Error (MSE)

Error terhadap data kita kuadratkan lalu dirata-ratakan. Semakin kecil semakin baik.

### Root Mean Squared Error (RMSE)

akar dari MSE

### Accuracy

\begin{equation}
Accuracy = \frac{TP+TN}{(TP+FP + TN+FN)}
\end{equation}

Berapa yang benar dibandingkan semua data

### Precision

\begin{equation}
Precision = \frac{TP}{(TP+FP)}
\end{equation}

- Target: FP sekecil mungkin
- Kapan dipakai? Data negatifnya sedikit, kita perlu lebih presisi mana yang negatif.
- Contoh:
    - klasifikasi apakah pakaian dicuci dengan mesin cuci atau non mesin cuci (dry clean dsb)
    - kebanyakan label kita adalah mesin cuci, beberapa pakaian yang tidak boleh menggunakan mesin cuci
    - Akan fatal kalau pakaian yang dry clean kita cuci dengan mesin cuci (False Positive)

### Recall / Sensitivity

\begin{equation}
Recall = \frac{TP}{(TP+FN)}
\end{equation}

- Target: FN sekecil mungkin
- Kapan dipakai? Data positifnya sedikit, kita perlu lebih sensitif mana yang positif.
- Contoh:
    - klasifikasi email spam vs non spam
    - kebanyakan email kita adalah non spam, beberapa email yang sebenarnya adalah spam
    - Akan fatal kalau email spam kita anggap non spam (False Negative)

### F1-Score

\begin{equation}
\frac{1}{F_1} = \frac{1}{2} \left(\frac{1}{Precision} + \frac{1}{Recall}\right)
\end{equation}

- rata-rata harmonik dari precision dan recall
- Kapan dipakai? Kalau mau balance antara precision dan recall
